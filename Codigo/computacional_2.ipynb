## 1. Correlación de los Datos

|    |      |      |      |      |      |      |      |      |      |      |      |
|----|------|------|------|------|------|------|------|------|------|------|------|
| X  | 2,00 | 2,10 | 2,20 | 2,30 | 2,40 | 2,50 | 2,60 | 2,70 | 2,80 | 2,90 | 3,00 |
| Y  | 2,78 | 3,29 | 3,29 | 3,33 | 3,23 | 3,69 | 3,46 | 3,87 | 3,62 | 3,40 | 3,99 |

```python
x_vals = np.array([2.00, 2.10, 2.20, 2.30, 2.40, 2.50, 2.60, 2.70, 2.80, 2.90, 3.00])
y_vals = np.array([2.78, 3.29, 3.29, 3.33, 3.23, 3.69, 3.46, 3.87, 3.62, 3.40, 3.99])
correlation = np.corrcoef(x_vals, y_vals)[0, 1]
lin_regress_result = linregress(x_vals, y_vals)
pendiente = lin_regress_result.slope
ordenada = lin_regress_result.intercept
r_cuadrado = lin_regress_result.rvalue**2


fig, ax = plt.subplots(figsize=(8, 4))
ax.errorbar(x_vals, y_vals, yerr=0.3, fmt="o", capsize=3, ecolor="k", markeredgecolor="k", label=f"Correlación: {correlation:.3n}")
ax.plot([0, 5], [ordenada, pendiente * 5 + ordenada], "C3-", lw=1, label="Ajuste Lineal")
ax.grid(True)
ax.legend()
ax.set_xlim(0, 5)
ax.set_xlabel("$X$")
ax.set_xticks(ax.get_xticks())
ax.set_xticklabels([f"{tick:.2n}" for tick in ax.get_xticks()])
ax.set_ylabel("$Y$")
ax.set_yticks(ax.get_yticks())
ax.set_yticklabels([f"{tick:.2n}" for tick in ax.get_yticks()])
plt.show()
```
