```python
import numpy as np
np.random.seed(42)
from scipy.stats import binom, nbinom, poisson, hypergeom

import matplotlib.pyplot as plt
import seaborn as sns
%config InlineBackend.figure_format = 'retina'
plt.rcParams['font.family'] = 'serif'
plt.rcParams['mathtext.fontset'] = 'cm'
plt.rcParams['figure.dpi'] = 120
plt.rcParams['legend.fontsize'] = "small"
plt.rcParams['axes.labelsize'] = 'large'
```

```python
mean = 30
prob = 0.2


dist_info = {
    "Binomial": {"dist": binom(mean/prob, prob), "color": "darkorange"},
    "Negative Binomial": {"dist": nbinom(mean*prob/(1-prob), prob), "color": "brown"},
    "Hypergeometric": {"dist": hypergeom(1000, 1000*mean/100, 100), "color": "red"},
    "Poisson": {"dist": poisson(mean), "color": "green"},
}

fig, ax = plt.subplots(1, 1, figsize=(8, 4))
n_vals = np.arange(0, 100+1, 1, dtype=np.int32)
for func_name, dist_dict in dist_info.items():
    prob_vals = dist_dict["dist"].pmf(n_vals)
    ax.plot(
        n_vals,
        prob_vals,
        "o--",
        c=dist_dict["color"],
        ms=4,
        mew="0.5",
        mec="k",
        alpha=0.75,
        label=func_name,
    )
ax.grid()
ax.legend()
ax.set_ylim(bottom=0)
ax.set_xlim(left=0, right=max(n_vals))
plt.show()
```

```python
# 3:
suman_siete = 0
total_casos = 0
for d1 in range(1, 7):
    for d2 in range(1, 7):
        if d1 + d2 == 7:
            suman_siete += 1
        total_casos += 1
prob_de_siete = suman_siete / total_casos
for n in range(55, 70+1):
    k_vals = np.arange(0, n+1, 1, dtype=np.int32)
    prob_vals = binom.pmf(k_vals, n, prob_de_siete)
    i_max = np.argmax(prob_vals)
    k_max = k_vals[i_max]
    print(f"Para {n = }, se tiene que la moda es {k_max = }")
    print(f"Diff con vecinos: {np.diff(prob_vals[i_max-1:i_max+2])}")
```
