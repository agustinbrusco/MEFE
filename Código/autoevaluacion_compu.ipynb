# $\text{V.}$
hacer un histograma (con las cajitas de ancho 0.05) de los números que pasaron la selección. Ajusta el rango del eje de las absisas, para que corra de -0.5 hasta 2.0.

```python
vals = nums[mayor_que_075]

fig, ax = plt.subplots(1, 1)
sns.histplot(vals, binwidth=0.05, ax=ax)
ax.set_xlim(-0.5, 2.0)
plt.show()
```

# $\text{VI.}$
superponer al histograma la función $f = n \times \alpha$, donde $\alpha = N ∗ 0.75 ∗ 0.05$

```python
alpha = N*0.75*0.05
n_vals = np.linspace(-0.5, 2.0, 100)
# Interpreté n como la variable independiente y no como el n de antes.
f_vals = n_vals * alpha
fig, ax = plt.subplots(1, 1)
sns.histplot(vals, binwidth=0.05, ax=ax, zorder=4)
ax.plot(n_vals, f_vals, "C1", zorder=5)
ax.grid()
ax.set_xlim(-0.5, 2.0)
ax.set_ylim(bottom=0)
plt.show()
```

# $\text{VII.}$
para cada bin del histograma, medir su distancia a $f$, y hacer el promedio de estas distancias

# Solución de BingChat:

```python
import random
import matplotlib.pyplot as plt

N = 1000
numeros_seleccionados = []

for i in range(N):
    numero = random.uniform(0, 1)
    if numero < 0.75:
        numeros_seleccionados.append(numero)

n = len(numeros_seleccionados)
plt.hist(numeros_seleccionados, bins=[x * 0.05 for x in range(-10, 41)])
plt.xlim(-0.5, 2.0)

alpha = N * 0.75 * 0.05
x = [x * 0.05 for x in range(-10, 41)]
y = [n * alpha for _ in x]  # a: Bing interpreted `f` as a constant.
plt.plot(x, y)

distancias = []
for i in range(len(x)):
    distancias.append(abs(y[i] - n))
distancia_promedio = sum(distancias) / len(distancias)
```
