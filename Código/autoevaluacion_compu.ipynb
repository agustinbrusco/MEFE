```python
import numpy as np
np.random.seed(42)

import matplotlib.pyplot as plt
import seaborn as sns
%config InlineBackend.figure_format = 'retina'
plt.rcParams['font.family'] = 'serif'
plt.rcParams['mathtext.fontset'] = 'cm'
plt.rcParams['figure.dpi'] = 120
plt.rcParams['legend.fontsize'] = "small"
plt.rcParams['axes.labelsize'] = 'large'
```

# $\text{I.}$
tomar un número al azar (equiprobable) entre 0. y 1.

```python
num = np.random.random()
print(f"{num = }")
```

# $\text{II.}$
fijarse si es menor que 0.75

```python
print(f"{num < 0.75 = }")
```

# $\text{III.}$
repetir los dos pasos anteriores N=1000 veces

```python
N = 1_000
nums = np.random.random(N)
mayor_que_075 = nums < 0.75
```

# $\text{IV.}$
¿cuántos números (n) pasaron esta selección?

```python
n = sum(mayor_que_075)
print(f"{n = }")
```

# $\text{V.}$
hacer un histograma (con las cajitas de ancho 0.05) de los números que pasaron la selección. Ajusta el rango del eje de las absisas, para que corra de -0.5 hasta 2.0.

```python
vals = nums[mayor_que_075]

fig, ax = plt.subplots(1, 1)
sns.histplot(vals, binwidth=0.05, ax=ax)
ax.set_xlim(-0.5, 2.0)
plt.show()
```

# $\text{VI.}$
superponer al histograma la función $f = n \times \alpha$, donde $\alpha = N ∗ 0.75 ∗ 0.05$

```python

alpha = N*0.75*0.05
n_vals = np.linspace(-0.5, 2.0, 100)
f_vals = n_vals * alpha
fig, ax = plt.subplots(1, 1)
sns.histplot(vals, binwidth=0.05, ax=ax)
ax.plot(n_vals, f_vals)
ax.set_xlim(-0.5, 2.0)
plt.show()
```
